```python
from PIL import Image
import numpy as np
import scipy
from scipy.signal import fftconvolve
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Make data.
X = np.arange(-5, 5, 0.25)
xlen = len(X)
Y = np.arange(-5, 5, 0.25)
ylen = len(Y)
X, Y = np.meshgrid(X, Y)
Z = np.sin(X*2*np.pi) + np.sin(Y*2*np.pi)
plt.imshow(Z)
```

```python
imfft = np.fft.fft2(Z)
imfft_re = np.fft.fftshift(imfft)
immag = np.abs(imfft_re)
plt.imshow(immag)
```
