```python
from PIL import Image
import numpy as np
import scipy
from scipy.signal import fftconvolve
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#real image
```

```python
im = np.array(Image.open('/Users/shjiao/Desktop/half-portrait-photo.jpg'))

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])


im_2= rgb2gray(im)
plt.imshow(im_2)
```

```python
 #do DFT(fft2) and plot MAGNITUDE spectrum
```

```python
imfft = np.fft.fft2(im_2)
imfft_re = np.fft.fftshift(imfft)
immag = np.log10(np.abs(imfft_re))
plt.imshow(immag)
```

```python
#plot PHASE spectrum
```

```python
imphase = np.log10(((np.angle(imfft_re)*180./np.pi)))
plt.imshow(imphase)
```
